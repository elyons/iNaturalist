```python
import requests
import pandas as pd

taxon_id = 48662
finished = 0
page = 1
output = ""
#r = requests.get("https://www.inaturalist.org/observations.csv?taxon_id=48662&on=2016-01&per_page=5&page="+str(page))
#r.text

while finished == 0 :
    r = requests.get("https://www.inaturalist.org/observations.csv?taxon_id="+str(taxon_id)+"&on=2016-01&per_page=5&page="+str(page))
    if len(r.text.split('\n')) < 3 :
        finished = 1
    if page == 1 :
        output = r.text
    else :
        output = output + r.text.split('\n', 1)[-1]
    print ("Finished page: ", page)
    page = page + 1
   
        
```
